```python
import pandas as pd
df = pd.read_csv('amazon_orders.csv')
df.head(10)
```

```python
df=  df.isnull()
df.head(10)
```

```python
df= df.fillna(0)
df.head(10)
```

```python
df.head()
```

```python
df['Total Charged'] = df['Total Charged'].str.replace('$','').astype('float')
df.head()
```

```python
type('Total Charged')
```

```python
df['Total Charged'].sum()
```

```python
df['Total Charged'].mean()
```

```python
df['Total Charged'].median()
```

```python
df['Total Charged'].max()
```

```python
df['Total Charged'].min()
```

```python
df['Tax Charged'] = df['Tax Charged'].str.replace('$','').astype('float')
df.head()
```

```python
df['Tax Charged'].sum() #Total tax paid
```

```python
df['Tax Charged'].sum()/df['Total Charged'].sum() #tax paid percentage
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```
